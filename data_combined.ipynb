```python
# Audio files

import os
import librosa

DATASET_PATH = "Data/genres_original"

X, y = [], []

for genre in os.listdir(DATASET_PATH):
    genre_path = os.path.join(DATASET_PATH, genre)

    if not os.path.isdir(genre_path):
        continue

    for file in os.listdir(genre_path):
        if file.endswith(".wav"):
            file_path = os.path.join(genre_path, file)

            try:
                audio, sr = librosa.load(file_path, sr=None)
                X.append(audio)
                y.append(genre)
            except Exception as e:
                print("Skipping:", file_path)

print("Loaded", len(X), "files")
```

```python
#Combining Images and Audio Files

import os
import numpy as np
from PIL import Image
import librosa

BASE_PATH = "/Users/shaveen/Desktop/ds6050/Data"

AUDIO_PATH = os.path.join(BASE_PATH, "genres_original")
IMAGE_PATH = os.path.join(BASE_PATH, "images_original")

X_audio = []
X_image = []
y = []

for genre in os.listdir(AUDIO_PATH):
    audio_genre_path = os.path.join(AUDIO_PATH, genre)
    image_genre_path = os.path.join(IMAGE_PATH, genre)

    if not os.path.isdir(audio_genre_path):
        continue

    for file in os.listdir(audio_genre_path):
        if file.endswith(".wav"):
            base_name = file.replace(".wav", "")
            image_name = base_name.replace(".", "")

            audio_file = os.path.join(audio_genre_path, file)
            image_file = os.path.join(image_genre_path, image_name + ".png")

            if not os.path.exists(image_file):
                continue

            try:
                audio, sr = librosa.load(audio_file, sr=None)

                img = Image.open(image_file).convert("RGB")
                img = img.resize((128, 128))

                X_audio.append(audio)
                X_image.append(np.array(img))
                y.append(genre)

            except Exception as e:
                print("Skipping:", file)

print("Loaded:", len(X_audio), "paired samples")
```
